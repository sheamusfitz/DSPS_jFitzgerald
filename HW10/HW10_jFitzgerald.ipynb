# 1: Brief Data Exploration

```
print(genes.mean(),'\n', "{:0.4f}±{:0.4f}".format(genes.mean().mean(),
                                                  genes.mean().std()),
      '\n\n', genes.std(), '\n', "{:0.4f}±{:0.4f}".format(genes.std().mean(),
                                                          genes.std().std()))
```
